```python
from keras.layers import Dense
from keras.models import Sequential


```

## Regression problem 

```python
df=pd.read_csv("carPrice_cleanedDf.csv")
df.head()
```

```python
df=df.drop(columns=['car_ID'])
X=df.drop(columns='price')
y=df.price
        
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=2023)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
y_train
```

```python
#setup the model architecture

model=Sequential()

#add the first layer

model.add(Dense(units=500,activation='relu',input_shape=(24,),kernel_initializer='he_uniform'))

# add the second layer
model.add(Dense(units=100,activation='relu',kernel_initializer='he_uniform'))
#add the output layer
model.add(Dense(units=1))
```

```python
#complile the model

model.compile(optimizer='sgd',loss='mean_squared_logarithmic_error')
```

```python
from keras.callbacks import EarlyStopping

#earlystopping is used when we don't know how many epochs needs to be used to optimize the model so early stopping will stop after sertain epochs when there is no
#imporvment in validation score
early_stopping_monitor=EarlyStopping(patience=2)


model.fit(X_train,y_train,validation_split=0.33,epochs=500,callbacks=[early_stopping_monitor])
```
