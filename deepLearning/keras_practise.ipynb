## Regression problem 

```python
df=pd.read_csv("carPrice_cleanedDf.csv")
df.head()
```

```python
df=df.drop(columns=['car_ID'])
X=df.drop(columns='price')
y=df.price
        
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=2023)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
y_train
```

```python
#setup the model architecture

model=Sequential()

#add the first layer

model.add(Dense(units=500,activation='relu',input_shape=(24,),kernel_initializer='he_uniform'))

# add the second layer
model.add(Dense(units=100,activation='relu',kernel_initializer='he_uniform'))
#add the output layer
model.add(Dense(units=1))
```

```python
#complile the model

model.compile(optimizer='sgd',loss='mean_squared_logarithmic_error')
```

```python
from keras.callbacks import EarlyStopping

#earlystopping is used when we don't know how many epochs needs to be used to optimize the model so early stopping will stop after sertain epochs when there is no
#imporvment in validation score
early_stopping_monitor=EarlyStopping(patience=2)


model.fit(X_train,y_train,validation_split=0.33,epochs=500,callbacks=[early_stopping_monitor])
```

```python
#load dataset-  Drug Df
import os
df_drug=pd.read_csv('drug_df_cleaned.csv')
```

```python
df_drug.head()
```

```python
df_drug.Drug.value_counts()
```

```python
#label encoding for target variable
from sklearn.preprocessing import LabelEncoder
X=df_drug.drop(axis=1,labels='Drug')
y=df_drug['Drug']
```

```python
labelencoder= LabelEncoder()
y=labelencoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2025)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
model=Sequential()
#input layer
model.add(Dense(units=100,activation='relu',input_dim=7))
#output layer
model.add(Dense(units=5,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam')
```

```python
model.fit(X_train,y_train,epochs=500,validation_split= 0.2)
```

```python
y_prob= model.predict(X_test)
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
