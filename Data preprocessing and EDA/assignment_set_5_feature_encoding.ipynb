### Q1. One hot encode the Type column and append it to the df

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.dropna(inplace=True)
df= pd.get_dummies(df,columns=['Type'],prefix=['Type_'])
```

### Q2. Encode the AirBags column in the df using frequency encoding

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
encode_airbags=df['AirBags'].map(df.groupby("AirBags").size()/len(df['AirBags']))
```

```python
df.loc[:,'AirBags_encoded']=encode_airbags
df
```

### Q3. Encode the DriveTrain column in the df using target encoding (consider Price as the target variable)

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df
```

```python
from category_encoders import target_encoder
target_encdoing_df= df.copy()
target_encdoing_df.groupby('DriveTrain')['Price'].count()
target_mean= target_encdoing_df['Price'].mean()
```

```python
target_encdoing_df.loc[:,"Encoded_DriveTrain"]=target_encdoing_df['DriveTrain'].map(target_encdoing_df.groupby('DriveTrain')['Price'].count()/target_mean)
```
