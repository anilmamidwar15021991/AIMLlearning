### Q1. Compute the mean, median and mode of all the numerical columns

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.head()
df.describe()
```

```python
for columnname in df.select_dtypes(include=np.number):
    print(f'column name: {columnname}')
    print(f'mean is {df[columnname].mean()}, median is {df[columnname].median()} ,mode is {df[columnname].value_counts().tolist()[0]}')
    
    
```

```python
#solution
#function to get mean, median and mode of all numeric coulmns of a DF
def centralTendancy(df):
    for columnname in df.select_dtypes(include=np.number):
        print(f'column name: {columnname}')
        print(f'mean is {df[columnname].mean()}, median is {df[columnname].median()} ,mode is {df[columnname].value_counts().tolist()[0]}')
    
```

```python
centralTendancy(df)
```

### 2. Compute the variance and IQR of all the numerical columns

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.head()
df.describe()
```

```python
#solution
#function to get variance, IQR of all numeric coulmns of a DF
def Dispersion(df):
    for columnname in df.select_dtypes(include=np.number):
        print(f'column name: {columnname}')
        Q3= np.percentile(df[columnname],75,interpolation='midpoint')
        Q1=np.percentile(df[columnname],25,interpolation='midpoint')                 
        print(f'variance is {df[columnname].var()}, IQF is {Q3-Q1}')
    
```

```python
Dispersion(df)
```

### 3. Compute the entropy of all the categorical columns

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.head()
```

```python
from math import log, e
```

```python
def pandas_entropy(column,base=None):
    vc=pd.Series(column).value_counts(normalize=True,sort=False)
    base = e if base is None else base
    return -(vc * np.log(vc)/np.log(base)).sum()
    
```

```python
for columnN in df.select_dtypes(exclude=np.number):
    entropy = pandas_entropy(df[columnN])
    print (f' entropy of {columnN} is {entropy}')
     
```
