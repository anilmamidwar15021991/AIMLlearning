### Q1. Find the pearson's correlation of all the numerical columns with Price column

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.dropna(inplace = True)

#move price column to the last
temp_cols= df.columns.tolist()
new_cols_position= temp_cols[0:4]+temp_cols[5:]+temp_cols[4:5]
df=df[new_cols_position]
df.head()
```

```python
from scipy.stats import pearsonr

#calcaulting relation of all linear variable with price column using pearsonr method

def pearsonrCorr(x,Y):
    print('x column is ' + x)
    corr,_= pearsonr(df[x],df[Y])
    print(f' pearson correlation is {corr:.5f} and p-value is {_:.23f}')
    
    
```

```python
for col in df.select_dtypes(include=np.number):
    pearsonrCorr(col,'Price')
    
```

### Q2. Find all the statistically significane numerical columns affecting the Price of vehicle (use spearmanr correlation & significance level 0.05)

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.dropna(inplace = True)
df.head()
```

```python
from scipy.stats import spearmanr
def corr_spearmanr(x,Y):
    corr, _ = spearmanr(df[x],df[Y])
    print(f' spearman correlation is  {corr:.3f} and p-value is  {_}')
          
          

for col in df.select_dtypes(include=np.number):
        print(col)
        corr_spearmanr(col,'Price')
    
```

### Q4. Find all the statistically significane numerical columns affecting the survival rate of tha passenger (use ANOVA & significance level 0.05)

```python
# Input
import scipy.stats as stats
import matplotlib.pyplot as plt 

df.dropna(inplace = True)
df.head()
```

```python
#tke the ramdom sample for each group

sample_df = df[['Survived','Age']]
unique_survived = df['Survived'].unique()
for survied in unique_survived:
    stats.probplot(df[df['Survived']== survied]['Age'],plot=plt ,dist='norm')
    
    plt.show()
```

```python
for survied in unique_survived:
    stats.probplot(df[df['Survived']==survied]['Parch'],plot=plt,dist='norm')
    
    plt.show()

```

```python
#above Q-Q plot shows that only age variable samples has normal probability distribution
```

```python
#we will check the anova significance using statsmodel api
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
from statsmodels.formula.api import ols
```

```python
for numerical_col in df.select_dtypes(include=np.number):
    model=ols(numerical_col + '~ C(Survived)',df).fit()
    print(f' Overall model F for column {numerical_col} is ({model.df_model:.0f}, {model.df_resid:.0f})={model.fvalue:.3f}, p={model.f_pvalue:.4f}')
    res=sm.stats.anova_lm(model,typ=2)
res
```
