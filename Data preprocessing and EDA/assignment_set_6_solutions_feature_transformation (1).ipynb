# Assignment Set 6 Solutions - Feature Scaling

```python
import pandas as pd
import numpy as np
```

### Q1. Standardize the Price column in Cars93 data

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
```

```python
# Solution 1

mean_value = df['Price'].mean()
std_value = df['Price'].std()

# Standardization
df['Price_standardized'] = (df['Price'] - mean_value)/std_value
df.head()
```

```python
# Solution 2

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df['Price_standardized'] = scaler.fit_transform(df['Price'].values.reshape(-1, 1))
df.head()
```

### Q2. Scale the Weight column in Cars93 data using robust scaling

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
```

```python
# Solution 1

median_value = df['Weight'].median()
q1 = np.percentile(df['Weight'], 25, interpolation = 'midpoint') 
q3 = np.percentile(df['Weight'], 75, interpolation = 'midpoint') 
IQR = q3 - q1 

# Standardization
df['Weight_standardized'] = (df['Weight'] - median_value)/IQR
df.head()
```

```python
# Solution 2

from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

df['Weight_standardized'] = scaler.fit_transform(df['Weight'].values.reshape(-1, 1))
df.head()
```

### Q3. Scale the Weight column in Cars93 data using min max scaling

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
```

```python
# Solution 1

min_value = df['Weight'].min()
max_value = df['Weight'].max()

# Standardization
df['Weight_standardized'] = (df['Weight'] - min_value)/(max_value - min_value)
df.head()
```

```python
# Solution 2

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df['Weight_standardized'] = scaler.fit_transform(df['Weight'].values.reshape(-1, 1))
df.head()
```
