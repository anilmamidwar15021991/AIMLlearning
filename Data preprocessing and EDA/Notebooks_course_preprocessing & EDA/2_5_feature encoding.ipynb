</div>

In Label Encoding, the categories in categorical feature are converted to the numeric forms. Primarily as per the alphabatical order. Once it's converted to numeric form, the feature can be used by the machine learning model.

**Note: It's advisable to use label encoding for target class only**

### Label Encoding using `cat.codes` function

Let's encode the 'label' feature in iris dataset



````python
import pandas as pd
iris_data = pd.read_csv('C:/Users/maste/Documents/Personal/ML+/Learning/Data preprocessing and EDA/Datasets_Course_Preprocessing_EDA/Iris.csv')
iris_data
````

````python
# Label Encoding using `cat.codes`function

iris_data['label'] = iris_data['label'].astype('category')
iris_data['label_encoded'] = iris_data['label'].cat.codes
iris_data[['label', 'label_encoded']].sample(10)
````

<div class="alert alert-info" style="background-color:#006666; color:white; padding:0px 10px; border-radius:5px;"><h2 style='margin:10px 5px'>3. One Hot Encoding</h2>
</div>

In One Hot Encoding, the categorical feature is removed and new binary features are added for every unique category in the categorical feature. 

In the above example of body style, 5 new features will be created corresponding to the 5 categories.

### One Hot Encoding using `get_dummies()` function from pandas

Let's check the dataframe.

````python
# Create data
one_hot_encoding_df = cat_data.copy()
one_hot_encoding_df.head()
````

````python
# one hot encoding
one_hot_encoding_df = pd.get_dummies(one_hot_encoding_df, columns=['body-style'], prefix=['body-style_'])
one_hot_encoding_df.head()
````

### One Hot Encoding using `LabelBinarizer` function from sklearn

````python
# Create data
one_hot_encoding_df = cat_data.copy()
one_hot_encoding_df.head()
````

````python
# Train label binarizer
from sklearn.preprocessing import LabelBinarizer

lb_encoder = LabelBinarizer()
ohe_feature = lb_encoder.fit_transform(one_hot_encoding_df['body-style'])
ohe_feature
````

Create dataframe from the array and concatenate with the master dataset

````python
# Concat with the master data
ohe_df = pd.DataFrame(ohe_feature, columns = lb_encoder.classes_)
ohe_df = ohe_df.add_prefix('body-style_')
one_hot_encoding_df = pd.concat([one_hot_encoding_df, ohe_df], axis=1).drop(['body-style'], axis=1)
one_hot_encoding_df.head()
````

**Note : One Hot Encoding works fine when we have less categories in the feature.**

### Limitations of One Hot Encoding

If the feature has lots of categories, one hot encoding will create one feature for every category. Hence the number of features will increase. So it's not advisable to use one hot encoding with lots of categories

<div class="alert alert-info" style="background-color:#006666; color:white; padding:0px 10px; border-radius:5px;"><h2 style='margin:10px 5px'>4. Frequency Encoding</h2>
</div>

In Frequency Encoding, the categories are replaced with their frequencies in the categorical feature.

In the above example of body style, frequency of all the 5 categories will be calculated and the categories will be replaced with the frequency.

````python
# Create data
freq_encoding_df = cat_data.copy()
freq_encoding_df.head()
````

````python
# Frequency encoding

fq = freq_encoding_df.groupby('body-style').size()/len(freq_encoding_df)   
freq_encoding_df['body-style'].map(fq)
````

````python
# mapping values to dataframe
freq_encoding_df.loc[:, "{}_freq_encode".format('body-style')] = freq_encoding_df['body-style'].map(fq)  
freq_encoding_df.head()
````

Let's cross check how are these values encoded

````python
freq_encoding_df[['body-style', 'body-style_freq_encode']].drop_duplicates()
````

**Note : Frequency Encoding works fine when you have nominal features.**

### Limitations of Frequency Encoding

In frequency encoding, the numeric values are assigned as per the frequency of the categories. So the machine learning model will treat it like a numerical value. It will treat 0.3 as greater than 0.2, 0.2 greater than 0.1 and so on. While such relationship doesn't holds true for the categories. 

It might generate the priority issue while training the model.

</div>

Ordinal Encoding is slightly different from the label encoding as it retains the ordinal characterstics of the feature. 

Let's create a height feature with tall, short, dwarf categories and encode it.

### Ordinal Encoding using `OrdinalEncoder` function from sklearn

Numeric values are assigned to the categories of a categorical feature as per the order that is observed in the feature. However, if you wish to define as specific order, you can specify it using "categories" parameter.

````python
import numpy as np
import pandas as pd
````

````python
ordinal_encoding_df = pd.DataFrame({'Height': ['tall', 'short', 'dwarf', 'dwarf', 'short', 'short'],
                                  'Weight': [50, 40, 30, 35, 45, 43]})
ordinal_encoding_df.head()
````

````python
# ordinal encoding
from sklearn.preprocessing import OrdinalEncoder

# define ordinal encoding
encoder = OrdinalEncoder()
````

````python
ordinal_encoding_df['Height'].to_numpy().reshape(-1,1)
````

````python
# Fit
encoder.fit(ordinal_encoding_df['Height'].to_numpy().reshape(-1,1))
encoder.categories_
````

````python
# Transform
ordinal_encoding_df['encoded_height'] = encoder.transform(ordinal_encoding_df['Height'].to_numpy().reshape(-1,1))
ordinal_encoding_df.head(15)
````

__Explicitly set order of categories --> Reverse the order__

````python
cats = [np.array(['tall', 'short', 'dwarf'], dtype=object)]
encoder = OrdinalEncoder(categories=cats, handle_unknown='use_encoded_value', unknown_value=-1)
````

````python
# Transform data
ordinal_encoding_df['encoded_height'] = encoder.fit_transform(ordinal_encoding_df['Height'].to_numpy().reshape(-1,1))
ordinal_encoding_df.head(15)
````

# Blending method

````python
import pandas as pd
df= pd.DataFrame({'x_0':['a']*5 +['b']*5,'x_1':['a']*9+['d']*1,'y':[1,1,1,1,0,1,0,0,0,0]})
df
````

````python
#mean of x_0 with respect to target variable y 
mean_x0=df.groupby('x_0')['y'].mean()
mean_x0
````

````python
#replace each value in x_0 with respetive category mean
df['x_0']=df['x_0'].map(mean_x0)
df
````

````python
#repeat the above steps for x_1
mean_x1=df.groupby('x_1')['y'].mean()
df['x_1']=df['x_1'].map(mean_x1)
df
````

#in the above dataset , we have 0 value for x_1 feature because of very few category values and this will add overfitting issue . to overcome this we will use target encoding with smoothing

we will write a function to smooth the taregt encoding


````python
def calc_smooth_mean(df,by,on,m):
    #compute the global mean
    mean= df[on].mean()
    
    #compute the number of obeservations and the mean of each group    
    agg=df.groupby(by)[on].agg(['count','mean'])
    counts=agg['count']
    means=agg['mean']
    
    #compute the smoothed mean
    smooth = (counts* means + m* mean)/(counts+m)
    print (smooth)
    return df[by].map(smooth)
    
````

````python
df['x_0']=calc_smooth_mean(df,by='x_0',on='y',m=10)
df['x_1']=calc_smooth_mean(df,by='x_1',on='y',m=10)
````
