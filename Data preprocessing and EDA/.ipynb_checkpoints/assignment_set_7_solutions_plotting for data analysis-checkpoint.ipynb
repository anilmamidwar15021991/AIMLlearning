### Q1. Plot the Air Passenger traffic change over time

```python
# Input
df = pd.read_csv('https://github.com/selva86/datasets/raw/master/AirPassengers.csv')
df.head()
```

```python
# Solution

# Draw Plot
plt.figure(figsize=(16,10), dpi= 80)
plt.plot('date', 'value', data=df, color='tab:red')

# Decoration
plt.ylim(50, 750)
plt.xticks(np.arange(0, len(df), 12), rotation=45)
plt.yticks(fontsize=12, alpha=.7)
plt.title("Air Passengers Traffic", fontsize=22)
plt.grid(axis='both', alpha=.3)
   
plt.show()
```

### Q2. Plot the Air Passenger traffic change over time and plot all the the peaks and troughs and annotate the occurence of some special events (every 3rd peak and trough)

```python
# Input
df = pd.read_csv('https://github.com/selva86/datasets/raw/master/AirPassengers.csv')
```

```python
# Solution

# Get the Peaks and Troughs
data = df['value'].values
doublediff = np.diff(np.sign(np.diff(data)))
peak_locations = np.where(doublediff == -2)[0] + 1

doublediff2 = np.diff(np.sign(np.diff(-1*data)))
trough_locations = np.where(doublediff2 == -2)[0] + 1

# Draw Plot
plt.figure(figsize=(16,10), dpi= 80)
plt.plot('date', 'value', data=df, color='tab:blue', label='Air Traffic')
plt.scatter(df.date[peak_locations], df.value[peak_locations], marker=mpl.markers.CARETUPBASE, color='tab:green', s=100, label='Peaks')
plt.scatter(df.date[trough_locations], df.value[trough_locations], marker=mpl.markers.CARETDOWNBASE, color='tab:red', s=100, label='Troughs')

# Annotate
for t, p in zip(trough_locations[::3], peak_locations[::3]):
    plt.text(df.date[p], df.value[p]+15, df.date[p], horizontalalignment='center', color='darkgreen')
    plt.text(df.date[t], df.value[t]-35, df.date[t], horizontalalignment='center', color='darkred')

# Decoration
plt.ylim(50,750)
plt.xticks(np.arange(0, len(df), 12), rotation=45)
plt.title("Peak and Troughs of Air Passengers Traffic (1949 - 1969)", fontsize=22)
plt.yticks(fontsize=12, alpha=.7)


plt.legend(loc='upper left')
plt.show()
```

### Q3. Plot multiple time series of same range on the same chart
- plot mdeaths and fdeaths

```python
# Input
df = pd.read_csv('https://github.com/selva86/datasets/raw/master/mortality.csv')
df.head()
```

```python
# Solution
mycolors = ['tab:red', 'tab:blue']    

# Draw Plot and Annotate
fig, ax = plt.subplots(1,1,figsize=(16, 9), dpi= 80)    

columns = df.columns[1:]  
for i, column in enumerate(columns):    
    plt.plot(df.date.values, df[column].values, lw=1.5, color=mycolors[i])    
    plt.text(df.shape[0]+1, df[column].values[-1], column, fontsize=14, color=mycolors[i])


# Decorations    
plt.tick_params(axis="both", which="both", bottom=False, top=False,    
                labelbottom=True, left=False, right=False, labelleft=True)        

plt.xticks(np.arange(0, len(df), 6), rotation=45)
plt.title('Number of Deaths from Lung Diseases in the UK (1974-1979)', fontsize=22)        
plt.xlim(-2, 80)    
plt.show()
```

### Q4. Compare the number of unique vehicles manufactured under various vehicle types. Draw a bar plot to represent it and arrange the bars in descending order

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.head()
```

```python
# Solution

#Prepare Data
df = df.groupby('Type').size().reset_index(name='counts').sort_values('counts', ascending = False)
df.head()
```

Now let's draw the bar plot

```python
# Plot Bars
plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df['Type'], df['counts'], width=.5)

# Add the count on top of bar plot
for i, val in enumerate(df['counts'].values):
    plt.text(i, val, float(val), 
             horizontalalignment='center', verticalalignment='bottom', 
             fontdict={'fontweight':500, 'size':12})

# Decoration
plt.gca().set_xticklabels(df['Type'], rotation=60, horizontalalignment= 'right')
plt.title("Number of Vehicles by Type", fontsize=22)
plt.xlabel('Type', fontsize=12)
plt.ylabel('# Vehicles', fontsize=12)
plt.ylim(0, 45)
plt.show()
```

### Q5. Compare the number of unique vehicles under every vehicle type and group it by DriveTrain. Draw a stacked bar chart for it, group the charts by DriveTrain

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.head()
```

```python
# Stacked Bar chart
df_stacked = df.groupby('Type').DriveTrain.value_counts().unstack()
df_stacked.plot.bar(figsize=(16,10), stacked = True)

# You can add the element properties individually as well
plt.title("Number of Vehicles by Vehicle Type and DriveTrain", fontsize=22)
plt.xlabel('Type of Vehicle', fontsize=12)
plt.ylabel('# Vehicles', fontsize=12)
plt.ylim(0, 40)
plt.show()
```

### Q6.  Check the distribution of Price in Cars93 dataset

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.head()
```

### Q7.  Check the relation between MPG.highway and Weight of the vehicle using a regression plot

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.head()
```

```python
# Solution

# Plot Bars
plt.figure(figsize=(16,10), dpi= 100)
sns.regplot(df['Weight'], df['MPG.highway'])

plt.xticks(fontsize=8); plt.yticks(fontsize=8)
plt.title("Scatterplot of MPG Highway vs Weight", fontsize=14)   
plt.show()
```

### Q8.  Check the relation between MPG.highway and Weight of the vehicle using a regression plot. Check if different type of vehicles have different relation.

```python
# Input
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv')
df.head()
```

```python
# Solution

# Multiple Lines of best fit in same plot
sns.lmplot(data=df, x='MPG.highway', y='Weight', hue='Type', ci=None, height=8.27, aspect=11.7/8.27)

plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("Scatterplot of Midwest Area vs Population", fontsize=24)   
plt.show();
```
